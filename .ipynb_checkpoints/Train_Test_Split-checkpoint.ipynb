```python
import pandas as pd 
import numpy as nm 
import matplotlib.pyplot as mlt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
# x= Feature 
# Y = Target 
```

```python
 data = pd.read_csv("titanic_data.csv") 
 data.head()
```

```python
# Je Col Dorkar nai delete
```

```python
data.drop(["PassengerId","Name","Ticket"],axis=1,inplace=True)
```

```python
data
```

```python
# 2 -> Feature and Target Col alada korre rakbo
```

```python
# Check Data Missing
data.isnull().sum()
```

```python
data.drop("Cabin",axis=1,inplace=True) 
data
```

```python
x = data.drop(["Survived"],axis=1) # Feature Col 
x
```

```python
y = data["Survived"] #Target Col
y
```

```python
# TRAIN / TEST / SPLIT 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Missing Value Fill -> Impute
```

```python
# 1. Delete Row
```

```python
# 1.1 Delete Col
```

```python
# 2. Impute Kono value diye Fill korre dauya 
# Numeric => Mean,Median 
# Categoric => Mode (Bessi sonka jaita oita), "Missing"
```

```python
data.isnull().sum()
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
# Impute -> Numeric 
## Joto kaj hobbe sob Train ar jonno
```

```python
# Mean Impute
```

```python
age_mean = X_train["Age"].mean() 
age_mean
```

```python
X_train["age_mean"] = X_train["Age"].fillna(age_mean) 
X_train
```

```python
X_test["age_mean"] = X_test["Age"].fillna(age_mean) 
X_test
```

```python
# Median Impute
```

```python
age_median = X_train["Age"].median() 
age_median
```

```python
X_train["age_median"] = X_train["Age"].fillna(age_median) 
X_train
```

```python
X_test["age_median"] = X_test["Age"].fillna(age_median) 
X_test
```

```python
sns.kdeplot(data=X_train,x=X_train.age_mean)
```

```python
sns.kdeplot(data=X_train,x=X_train.age_median)
```

```python
 # Imputation Though ML
```

```python
from sklearn.impute import SimpleImputer
```

```python
age_imp = SimpleImputer(missing_values=nm.nan,strategy="mean") 
age_imp.fit(X_train[["Age"]]) 
X_train["Age"] = age_imp.transform(X_train[["Age"]])
```

```python
X_train.isnull().sum()
```

```python
X_train.drop(["age_mean","age_median"],axis=1,inplace=True)
```

```python
X_train
```

```python
X_test["Age"] = age_imp.transform(X_test[["Age"]])
```

```python
X_test.head()
```

```python
X_test.drop(["age_mean","age_median"],axis=1,inplace=True)
```

```python
X_test.isnull().sum()
```

```python
# Category Value -> Mode / "Missing"
```

```python
data2 = pd.read_csv("titanic_data.csv") 
data2
```

```python
data2.isnull().sum()
```

```python
sns.countplot(data=data2,x=data2.Embarked)
```

```python
#Mode Value diye (Embared) impute korbo
```

```python
data2.drop(["PassengerId","Name","Ticket"],axis=1,inplace=True)
```

```python
embarked_emp = SimpleImputer(missing_values=nm.nan,strategy="most_frequent") 
embarked_emp.fit(data2[["Embarked"]]) 
data2["Embarked"] = embarked_emp.transform(data2[["Embarked"]]).ravel()
```

```python
data2.isnull().sum()
```

```python
# Missing
```

```python
cabin_emp = SimpleImputer(missing_values=nm.nan,strategy="constant",fill_value="Missing",add_indicator=True) 
cabin_emp.fit(data2[["Cabin"]]) 
data2[["Cabin","cabin_miss_ind"]] = cabin_emp.transform(data2[["Cabin"]])
```
