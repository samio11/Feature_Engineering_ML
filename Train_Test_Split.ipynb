```python
 # Outlier ->
# Detection -> 
# 1. Z-Score 
# 2. IQR 

# Detech ses a 
# 1. Delete
# 2. Cap (Limited range ar modde data nie asbo)
```

```python
# 1. Z-Score 
# Z = (data point - mean) / std 
# Z = -3 theke choto / 3 theke boro -> Outlier
```

```python
## Age Outlier
```

```python
mean_age = X_train["Age"].mean() 
std_age = X_train["Age"].std() 
print("Mean Value",mean_age) 
print("Standerd Deviasion",std_age)
X_train["zscore_age"] = (X_train["Age"] - mean_age) / std_age
```

```python
X_train.head()
```

```python
outliers = X_train[abs(X_train["zscore_age"]) > 3] 
print("OutLiers Found:-",len(outliers)) 
display(outliers)
```

```python
## Fare Outlier
```

```python
fare_mean = X_train["Fare"].mean() 
fare_std = X_train["Fare"].std() 
X_train["zscore_fare"] = (X_train["Fare"] - fare_mean) / fare_std
```

```python
X_train.head()
```

```python
outliers_fare = X_train[abs(X_train["zscore_fare"]) > 3] 
print(f"Fare Ourlier Detected {len(outliers_fare)}") 
display(outliers_fare)
```

```python
# IQR Method 
```

```python
# Age
```

```python
age_q1 = X_train["Age"].quantile(0.25) 
age_q3 = X_train["Age"].quantile(0.75) 
age_IQR = age_q3-age_q1 
age_min = age_q1 - 1.5*age_IQR 
age_max = age_q3 + 1.5*age_IQR 
print(f"Age Minimum:- {age_min}") 
print(f"Age Maximum:- {age_max}")
```

```python
age_outlier_IQR = X_train[(X_train["Age"]<age_min) | (X_train["Age"] > age_max)] 
print(f"Found Age IQR outlier:- {len(age_outlier_IQR)}") 
display(age_outlier_IQR)
```

```python
## Fare 
```

```python
fare_q1 = X_train["Fare"].quantile(0.25) 
fare_q3 = X_train["Fare"].quantile(0.75) 
fare_IQR = fare_q3 - fare_q1 
# fare_min = fare_q1 - 1.5*fare_IQR 
fare_min = max(0,fare_q1 - 1.5*fare_IQR) 
fare_max = fare_q3 + 1.5*fare_IQR 
print("Fare Min IQR:-",fare_min) 
print("Fare Max IRQ:-",fare_max)
```

```python
fare_outlier_IQR = X_train[(X_train["Fare"] < fare_min) | (X_train["Fare"]> fare_max)] 
print(f"Fare outlier detected:- {len(fare_outlier_IQR)}") 
display(fare_outlier_IQR)
```

```python
# Z-Scoring => Mean (Extreme sensative to Higher value / Normal Destribution 
```

```python
# IQR => Median value (No value for Higher value at all) / skewed left/right
```
